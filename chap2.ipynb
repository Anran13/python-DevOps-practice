```python
import re

line = '127.0.0.1 - rj [13/Nov/2019:14:34:30 -0000] "GET HTTP/1.0" 200'

r = r'(?P<IP>\d+\.\d+\.\d+\.\d+) - (?P<User>\w+) \[(?P<Time>\d{2}/\w{3}/\d{4}:\d{2}:\d{2}:\d{2} -\d{4})\] (?P<Request>\"\w{3} \w{4}/\d\.\d\") (?P<Status>\d{3})'

m = re.search(r, line)
print(m.group('IP'))    
print(m.group('User'))
print(m.group('Time'))  
print(m.group('Request'))
print(m.group('Status'))
```

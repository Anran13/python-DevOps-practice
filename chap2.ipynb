```python
import re

line = '127.0.0.1 - rj [13/Nov/2019:14:34:30 -0000] "GET HTTP/1.0" 200'

r = r'(?P<IP>\d+\.\d+\.\d+\.\d+) - (?P<User>\w+) \[(?P<Time>\d{2}/\w{3}/\d{4}:\d{2}:\d{2}:\d{2} -\d{4})\] (?P<Request>\"\w{3} \w{4}/\d\.\d\") (?P<Status>\d{3})'

m = re.search(r, line)
print(m.group('IP'))    
print(m.group('User'))
print(m.group('Time'))  
print(m.group('Request'))
print(m.group('Status'))
```

```python
# Symmetric key encryption using Advanced Encryption Algorithm (AEA)
from cryptography.fernet import Fernet
key = Fernet.generate_key()

# encrypt
f = Fernet(key)
message = b"Secrets go here"
encrypted_symmetric = f.encrypt(message)

# decrypt
f = Fernet(key)
f.decrypt(encrypted_symmetric)
```

```python
# Asymmetric key encryption using Rivest-Shamir-Adleman (RSA)
from cryptography.hazmat.backends import default_backend
from cryptography.hazmat.primitives.asymmetric import rsa
from cryptography.hazmat.primitives.asymmetric import padding
from cryptography.hazmat.primitives import hashes

private_key = rsa.generate_private_key(
    public_exponent=65537, 
    key_size=4096, 
    backend=default_backend()
    )
public_key = private_key.public_key()

message = b"More secrets go here"

# encrypt
encrypted_asymmetric = public_key.encrypt(
    message, 
    padding.OAEP(
        mgf=padding.MGF1(algorithm=hashes.SHA256()), 
        algorithm=hashes.SHA256(), 
        label=None)
    )

# decrypt
decrypted_asymmetric = private_key.decrypt(
    encrypted_asymmetric, 
    padding.OAEP(
        mgf=padding.MGF1(algorithm=hashes.SHA256()), 
        algorithm=hashes.SHA256(), 
        label=None)
    )

decrypted_asymmetric
```

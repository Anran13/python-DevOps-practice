```python
from numba import jit
import numpy as np

# 第一次呼叫 sum2d 時做 JIT 編譯，之後呼叫就直接用編譯後的機器碼執行，速度會快很多
@jit
def sum2d(arr):
    M, N = arr.shape
    result = 0.0
    for i in range(M):
        for j in range(N):
            result += arr[i, j]
    return result

a = np.arange(9).reshape(3, 3)
print(a)
print(sum2d(a))
```

```python
from numba import jit, njit
import numpy as np

# 在 nopython=True 模式下，若 Numba 不能完全編譯成機器碼，會直接報錯，能幫你確認是否真的在跑 JIT。  
@njit  # 或 @jit(nopython=True)
def sum_matrix_njit(M):
    N1, N2 = M.shape
    result = 0
    for i in range(N1):
        for j in range(N2):
            result += M[i, j]
    return result

x = np.arange(100).reshape(10, 10)
print(sum_matrix_njit(x))
```
